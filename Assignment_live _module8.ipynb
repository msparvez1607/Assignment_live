```python
#Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array 
import numpy as np
even=[]
odd=[]
#Extract all even and odd integers from array
a1=np.random.randint(1,10,(5,6))
a1_new=np.ravel(a1)
for items in a1_new:
    if items%2==0:
        even.append(items)
    else:
        odd.append(items)
print(f'Even numbers : {even}')
print(f'Odd numbers : {odd}')
```

```python
import pandas as pd
import numpy as np
date_range=pd.date_range(start='2024-01-01',periods=10,freq='D')
a,b=10,15
temp=a+(b-a)*np.random.rand(10)
hum=20+(10)*np.random.rand(10)
data=pd.DataFrame({'Date':date_range,'Temperature':temp,'Humidity':hum})
data.columns
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
ax1=plt.subplot()
ax1.plot(data['Date'],data['Temperature'],color='orange' )
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperatue')
ax1.tick_params(axis='y', labelcolor='orange')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(data['Date'], data['Humidity'], color='blue')
ax2.set_ylabel('Humidity', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')
plt.title('Plot of Temp. & Humidity on different days')
plt.legend

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate random data from a normal distribution
data = np.random.randn(100)  # 1000 data points

# Create a histogram of the data
plt.hist(data, bins=30, density=True, alpha=0.6, color='lightblue', edgecolor='black')

# Calculate mean and standard deviation of the data
mean = np.mean(data)
std = np.std(data)

# Create a range of values for the x-axis
x = np.linspace(min(data), max(data), 100)

# Compute the normal distribution curve (PDF)
pdf = norm.pdf(x, mean, std)

# Plot the normal distribution curve
plt.plot(x, pdf, 'r', linewidth=2, label='Normal Distribution (PDF)')

# Add labels and legend
plt.xlabel('Values')
plt.ylabel('Frequencies')
plt.legend()
plt.title('Histogram with PDF overlay')

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)

# Define function to classify points based on quadrant
def classify_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Quadrant I'
    elif x < 0 and y > 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    elif x > 0 and y < 0:
        return 'Quadrant IV'

# Create a DataFrame with the data
df = pd.DataFrame({'x': x, 'y': y})
df['Quadrant'] = df.apply(lambda row: classify_quadrant(row['x'], row['y']), axis=1)

# Create a Seaborn scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='Set1')

# Add labels, title, and legend
plt.axhline(0, color='gray', linestyle='--', linewidth=1)  # X-axis
plt.axvline(0, color='gray', linestyle='--', linewidth=1)  # Y-axis
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrant')
plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Ensure Bokeh outputs in the notebook (if using Jupyter)
output_notebook()

# Generate x values and compute y values for the sine wave
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave Function", x_axis_label='X', y_axis_label='sin(X)')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Show grid lines (default is True, but we ensure they are visible)
p.xgrid.visible = True
p.ygrid.visible = True

# Display the plot
show(p)
```
