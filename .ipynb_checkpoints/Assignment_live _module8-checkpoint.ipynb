```python
#Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array 
import numpy as np
even=[]
odd=[]
#Extract all even and odd integers from array
a1=np.random.randint(1,10,(5,6))
a1_new=np.ravel(a1)
for items in a1_new:
    if items%2==0:
        even.append(items)
    else:
        odd.append(items)
print(f'Even numbers : {even}')
print(f'Odd numbers : {odd}')
```

```python
#Create two NumPy arrays, x and y, each containing 100 random float values between 0 and 1. Perform the following tasks using Matplotlib and NumPy
import numpy as np
import matplotlib.pyplot as plt
x1=0.5
y1=0.5
x=np.random.rand(100)
y=np.random.rand(100)
plt.scatter(x,y,marker='o',color='red')
plt.axvline(x=0.5, color='blue', linestyle='--', label='x=0.5')
plt.axhline(y=0.5, color='green', linestyle='--', label='y=0.5')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Advanced Scatter plot of Random variables')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data for the scatter plot
x = np.random.rand(10)
y = np.random.rand(10)

# Create a scatter plot
plt.scatter(x, y, color="blue", label="Data Points")

# Add a line to the plot (e.g., y = x)
x_line = np.linspace(0, 1, 100)
y_line = x_line  # Line equation y = x
plt.plot(x_line, y_line, color="red", label="y = x")

# Add the specific point (x=0.5, y=0.5)
plt.scatter(0.5, 0.5, color="green", s=100, label="Point (0.5, 0.5)", marker='o')

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot with a Line and a Specific Point")

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
date_range=pd.date_range(start='2024-01-01',periods=10,freq='D')
a,b=10,15
temp=a+(b-a)*np.random.rand(10)
hum=20+(10)*np.random.rand(10)
data=pd.DataFrame({'Date':date_range,'Temperature':temp,'Humidity':hum})
data.columns
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
ax1=plt.subplot()
ax1.plot(data['Date'],data['Temperature'],color='orange' )
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperatue')
ax1.tick_params(axis='y', labelcolor='orange')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(data['Date'], data['Humidity'], color='blue')
ax2.set_ylabel('Humidity', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')
plt.title('Plot of Temp. & Humidity on different days')
plt.legend

```
